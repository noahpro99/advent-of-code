```python
file_name = "input.txt"
monkey_defs = [x.split("\n") for x in open(file_name).read().split("\n\n") if x]
monkeys = [
    m
    for m in [
        {
            "num_inspections": 0,
            "operation": x[2].split(" ")[6],
            "operand": x[2].split(" ")[7],
            "divisible_by": int(x[3].split(" ")[5]),
            "items": [int(i.strip(",")) for i in x[1].split(" ")[4:]],
            "true": int(x[4].split(" ")[9]),
            "false": int(x[5].split(" ")[9]),
        }
        for x in monkey_defs
    ]
]
operation = {
    "+": lambda x, y: x + y,
    "*": lambda x, y: x * y,
}
print(monkey_defs[0])
print(monkeys[0])
# product of divisible bys
prod = 1
for m in monkeys:
    prod *= m["divisible_by"]
print(prod)

for round in range(10000):
    print(round, [(m["num_inspections"]) / (round + 1) for m in monkeys])

    for monkey in monkeys:
        for item in monkey["items"]:
            try:
                new = (
                    operation[monkey["operation"]](item, int(monkey["operand"])) % prod
                )
            except:
                new = operation[monkey["operation"]](item, item) % prod
            if new % monkey["divisible_by"] == 0:
                monkeys[monkey["true"]]["items"].append(new)
            else:
                monkeys[monkey["false"]]["items"].append(new)
            monkey["num_inspections"] += 1
        monkey["items"] = []

inspections_list = [m["num_inspections"] for m in monkeys]
inspections_list.sort()
inspections_list, inspections_list[-1] * inspections_list[-2]
```
