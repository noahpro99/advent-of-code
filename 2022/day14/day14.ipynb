```python
file_name = "input.txt"
rock_paths_abs = [
    [[int(p) for p in i.split(",")] for i in x.split(" -> ")]
    for x in open(file_name).read().splitlines()
]
xs = [point[0] for points in rock_paths_abs for point in points]
ys = [point[1] for points in rock_paths_abs for point in points]
X0 = min(xs)
X0_FLOOR = -1000
rock_paths = [
    [
        [int(p) - X0 if idx == 0 else int(p) for idx, p in enumerate(i.split(","))]
        for i in x.split(" -> ")
    ]
    for x in open(file_name).read().splitlines()
]
rock_paths_floor = [
    [
        [
            int(p) - X0_FLOOR if idx == 0 else int(p)
            for idx, p in enumerate(i.split(","))
        ]
        for i in x.split(" -> ")
    ]
    for x in open(file_name).read().splitlines()
]

map = [["." for _ in range(max(ys) + 1)] for _ in range(max(xs) + 1 - X0)]
map_floor = [
    ["." for _ in range(max(ys) + 1 + 2)] for _ in range(max(xs) * 2 + 1 - X0_FLOOR)
]
print(max(xs) * 2 + 1 - X0_FLOOR, max(ys) + 1 + 2)


def print_map(map):
    for row in map:
        print("".join(row))


# print_map(map)

print(max(xs) - min(xs))
print(max(ys) - min(ys))


def add_paths(map, rock_paths):
    for rock_path in rock_paths:
        # print(rock_path)
        for i in range(len(rock_path) - 1):
            r1, r2 = rock_path[i], rock_path[i + 1]
            dx, dy = r2[0] - r1[0], r2[1] - r1[1]
            x, y = r1[0], r1[1]
            while abs(dx) != 0 or abs(dy) != 0:
                # print(x, y, dx, dy)
                map[x][y] = "#"
                if abs(dx) != 0:
                    x += 1 if dx > 0 else -1
                    dx += -1 if dx > 0 else 1
                if abs(dy) != 0:
                    y += 1 if dy > 0 else -1
                    dy += -1 if dy > 0 else 1
        map[rock_path[-1][0]][rock_path[-1][1]] = "#"


def add_floor(map):
    for i in range(len(map)):
        map[i][len(map[0]) - 1] = "#"


add_paths(map, rock_paths)
add_paths(map_floor, rock_paths_floor)
add_floor(map_floor)

# print_map(map)
```

```python
def drop_sand(map, source) -> tuple[int] | None:
    s = source.copy()
    while True:
        if map[s[0]][s[1] + 1] == ".":
            s[1] += 1
        elif s[0] - 1 < 0:
            return None
        elif map[s[0] - 1][s[1] + 1] == ".":
            s[0] -= 1
            s[1] += 1
        elif s[0] + 1 > len(map):
            return None
        elif map[s[0] + 1][s[1] + 1] == ".":
            s[0] += 1
            s[1] += 1
        else:
            return s


def max_sand(map, source) -> int:
    i = 0
    while True:
        s = drop_sand(map, source)
        if s is not None:
            map[s[0]][s[1]] = "o"
            i += 1
        else:
            return i


def max_sand_floor(map, source) -> int:
    i = 0
    while True:
        s = drop_sand(map, source)
        if s == source:
            return i+1
        if s is not None:
            map[s[0]][s[1]] = "o"
            i += 1
        else:
            return i


source = [500 - X0, 0]
source_floor = [500 - X0_FLOOR, 0]
ms = max_sand(map, source)
# print_map(map)
print(ms)
ms = max_sand_floor(map_floor, source_floor)
# print_map(map_floor)
print(ms)
```
