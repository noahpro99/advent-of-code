```python
file_name = "input.txt"
valves = {
    v.split(" ")[1]: {
        "rate": int(v.split(" ")[4][5:-1]),
        "leads": "".join(v.split(" ")[9:]).split(","),
    }
    for v in open(file_name).read().splitlines()
}
valves
```

```python
def most_pressure(valves: dict[str, dict], total_time: int, start: str):
    visited = {}
    most_pressure = 0
    frontier: list[tuple[int, int, str, set[str]]] = []
    frontier.append((0, 0, start, ()))
    i = 0
    while frontier:
        pressure, time, cur, opened = frontier.pop()
        i += 1
        if (cur, opened) in visited and time > visited[(cur, opened)]:
            continue
        visited[(cur, opened)] = time
        if time == total_time or len(opened) == len(valves):
            if pressure > most_pressure:
                most_pressure = pressure
            continue

        time += 1
        if cur not in opened and valves[cur]["rate"] > 0:
            frontier.append(
                (
                    pressure + (valves[cur]["rate"] * (total_time - time)),
                    time,
                    cur,
                    (*opened, cur),
                )
            )
        for option in valves[cur]["leads"]:
            frontier.append((pressure, time, option, opened))
    return most_pressure


mp = most_pressure(valves, 30, "AA")
print(mp)
```

```python
from heapq import heappop, heappush


def distances(valves):
    ds = {k: {} for k in valves.keys()}
    for k, v in ds.items():
        frontier = [(0, k)]
        visited = set()
        while frontier:
            d, cur = heappop(frontier)
            if cur in visited:
                continue
            visited.add(cur)
            if valves[cur]["rate"] > 0 and cur != k:
                v[cur] = d
            for option in valves[cur]["leads"]:
                heappush(frontier, (d + 1, option))

    return ds


worthwhile_valves_sorted = sorted(
    [v for v in valves.keys() if valves[v]["rate"] > 0],
    key=lambda x: valves[x]["rate"],
    reverse=True,
)
worthwhile_valves_sorted
ds = distances(valves)
ds, worthwhile_valves_sorted
```
