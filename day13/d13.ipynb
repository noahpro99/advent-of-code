```python

G = [[i for i in x] for x in open("input.txt", "r").read().splitlines()]
R = len(G)
C = len(G[0])
G
```

```python
def fall(G, dir):
    for r in range(R):
        for c in range(C):
            if not G[r][c] == "O":
                continue
            bottom = False
            rr = r
            cc = c
            while not bottom:
                if dir == "N":
                    above = rr - 1
                    if above < 0 or G[above][c] == "#":
                        bottom = True
                        break
                    else:
                        temp = G[rr][c]
                        G[rr][c] = G[above][c]
                        G[above][c] = temp
                        rr = above
                elif dir == "S":
                    below = rr + 1
                    if below >= R or G[below][c] == "#":
                        bottom = True
                        break
                    else:
                        temp = G[rr][c]
                        G[rr][c] = G[below][c]
                        G[below][c] = temp
                        rr = below
                elif dir == "E":
                    right = cc + 1
                    if right >= C or G[r][right] == "#":
                        bottom = True
                        break
                    else:
                        temp = G[r][cc]
                        G[r][cc] = G[r][right]
                        G[r][right] = temp
                        cc = right
                elif dir == "W":
                    left = cc - 1
                    if left < 0 or G[r][left] == "#":
                        bottom = True
                        break
                    else:
                        temp = G[r][cc]
                        G[r][cc] = G[r][left]
                        G[r][left] = temp
                        cc = left
    return G

fall(G, "N")
                        
```

```python
def stats(G):
    # num # num O num .
    return f"num #: {sum([x.count('#') for x in G])} num O: {sum([x.count('O') for x in G])} num .: {sum([x.count('.') for x in G])}"

def p2(G):
    p2 = 0
    for r in range(R):
        for c in range(C):
            if G[r][c] == "O":
                p2 += R-r
    return p2

x = []
dirs = ["N", "W", "S", "E"]
for i in range(int(200)):
    for dir in dirs:
        for j in range(200):
            G = fall(G, dir)
    x.append(p2(G))
```
